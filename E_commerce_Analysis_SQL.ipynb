#**E-Commerce Sales & Customer Behavior Analysis Using SQL**


##**Business Problem**

An online retail company has accumulated thousands of transaction records but lacks insight into its sales performance and customer purchasing behavior. Management wants to answer key questions such as:

+ Which products generate the most revenue?
+ Which countries contribute the highest sales?
+ Who are the most valuable customers?
+ Which months have the highest sales?
+ Are there products with high sales volume but low revenue?
+ How can the company improve profitability and customer retention?

##**Data Importing**

```python
# Import libraries
import pandas as pd
import sqlite3

# Load Excel file
df = pd.read_excel("Online_Retail.xlsx")

# Create SQLite database
conn = sqlite3.connect("ecommerce.db")

# Store table
df.to_sql("sales", conn, if_exists="replace", index=False)
```

```python
# Display the first 10 records
df_sample = pd.read_sql_query("SELECT * FROM sales LIMIT 10;", conn)
print(df_sample)
```

```python
# Count total transactions
df_transactions = pd.read_sql_query("SELECT COUNT(*) AS Total_Transactions FROM sales;", conn)
print(df_transactions)
```

```python
# Find missing Customer IDs
missing_customers = pd.read_sql_query("SELECT COUNT(*) AS Missing_Customers FROM sales WHERE CustomerID IS NULL;", conn)
print(missing_customer)
```

```python
# Calculate Total Revenue
total_revenue = pd.read_sql_query("SELECT SUM(Quantity * UnitPrice) AS Total_Revenue FROM sales;", conn)
print(total_revenue)
```

```python
# Top 10 Revenue-Producing Products
top_revenue_products = pd.read_sql_query("SELECT Description,SUM(Quantity * UnitPrice) AS Total_Revenue FROM sales GROUP BY Description ORDER BY Total_Revenue DESC LIMIT 10;", conn)
print(top_revenue_products)
```

```python
# Top 10 Customers by Spending
top_customers = pd.read_sql_query("SELECT CustomerID, SUM(Quantity * UnitPrice) AS Total_Spent FROM sales GROUP BY CustomerID ORDER BY Total_Spent DESC LIMIT 10;", conn)
print(top_customers)
```

```python
# Revenue by Country
revenue_by_country = pd.read_sql_query("SELECT Country, SUM(Quantity * UnitPrice) AS Total_Revenue FROM sales GROUP BY Country ORDER BY Total_Revenue DESC;", conn)
print(revenue_by_country)
```

```python
# Average Order Value
average_order_value = pd.read_sql_query("SELECT AVG(Quantity * UnitPrice) AS Average_Order_Value FROM sales;", conn)
print(average_order_value)
```

```python
# Monthly Revenue Analysis
monthly_revenue = pd.read_sql_query("SELECT strftime('%Y-%m', InvoiceDate) AS Month, SUM(Quantity * UnitPrice) AS Total_Revenue FROM sales GROUP BY Month ORDER BY Month;", conn)
print(monthly_revenue)
```

```python
# Customer Ranking
customer_ranking = pd.read_sql_query("SELECT CustomerID, SUM(Quantity * UnitPrice) AS Total_Spent, RANK() OVER(ORDER BY SUM(Quantity * UnitPrice) DESC) AS CustomerRank FROM sales GROUP BY CustomerID;", conn)
print(customer_ranking)
```

```python
# Products with high sales volume but low revenue
high_volume_low_revenue_products = pd.read_sql_query("""WITH ProductAggregates AS (
    SELECT
        Description,
        SUM(Quantity) AS TotalQuantity,
        SUM(Quantity * UnitPrice) AS TotalRevenue
    FROM
        sales
    GROUP BY
        Description
),
AverageMetrics AS (
    SELECT
        AVG(TotalQuantity) AS AvgQuantity,
        AVG(TotalRevenue) AS AvgRevenue
    FROM
        ProductAggregates
)
SELECT
    pa.Description,
    pa.TotalQuantity,
    pa.TotalRevenue
FROM
    ProductAggregates pa,
    AverageMetrics am
WHERE
    pa.TotalQuantity > am.AvgQuantity AND pa.TotalRevenue < am.AvgRevenue
ORDER BY
    pa.TotalQuantity DESC, pa.TotalRevenue ASC
LIMIT 10;""", conn)

print(high_volume_low_revenue_products)
```

```python
# Revenue Segmentation
revenue_segmentation = pd.read_sql_query("""SELECT
InvoiceNo,
Quantity * UnitPrice AS Revenue,
CASE
WHEN Quantity * UnitPrice < 100
THEN 'Low'
WHEN Quantity * UnitPrice BETWEEN 100 AND 500
THEN 'Medium'
ELSE 'High'
END AS RevenueCategory
FROM sales;""", conn)
print(revenue_segmentation)
```

```python
# Customers Above Average Spending
customers_above_avg_spending = pd.read_sql_query("""WITH CustomerTotals AS (
SELECT CustomerID, SUM(Quantity * UnitPrice) AS Spending
FROM sales
GROUP BY CustomerID
)
SELECT *
FROM CustomerTotals
WHERE Spending > (
SELECT AVG(Spending)
FROM CustomerTotals
);""", conn)
print(customers_above_avg_spending)
```

```python
# Create Revenue Summary View

conn.execute("""DROP VIEW IF EXISTS RevenueByCountry;""")
conn.execute("""CREATE VIEW RevenueByCountry AS
SELECT
Country,
ROUND(SUM(Quantity * UnitPrice),2) AS Revenue
FROM sales
GROUP BY Country;""")
print("VIEW 'RevenueByCountry' created successfully.")
```

```python
revenue_by_country_view = pd.read_sql_query("SELECT * FROM RevenueByCountry;", conn)
print(revenue_by_country_view)
```
